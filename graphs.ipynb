```python
import pickle
import matplotlib.pyplot as plt
```

```python
# Step 1: Load the pickle file
with open('results_with_time.pickle', 'rb') as file:
    data = pickle.load(file)

# Step 2: Extract time and result values
time_original_values = list(data.keys())
result_values = list(data.values())
```

```python
# Step 3: Plot the data
plt.plot(time_original_values, result_values)
plt.xlabel('Time (day)')
plt.ylabel('Luminosity')
plt.title('The scaled light curve')
plt.show()
```
